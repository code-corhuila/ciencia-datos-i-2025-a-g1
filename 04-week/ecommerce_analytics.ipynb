# Leer datos 
En esta sección del código, primero estamos importando las bibliotecas necesarias para trabajar con datos y visualizaciones. Luego, cargamos un archivo CSV que contiene datos de comercio electrónico en un DataFrame de pandas. Finalmente, mostramos las primeras filas del DataFrame para tener una vista previa de los datos que vamos a analizar.


````python
import pandas as pd

# Importar las librerías necesarias
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('data/ecommerce_data.csv')

# Mostrar el dataframe
df.head()
````

Después de haber leído y mostrado el DataFrame, podemos generar una gráfica que muestre la tendencia de los métodos de pago utilizados. Para ello, utilizaremos la biblioteca `matplotlib` para crear un gráfico de barras que represente la frecuencia de cada método de pago.

A continuación, se muestra el código para generar esta gráfica:


````python
# Contar la frecuencia de cada método de pago
payment_method_counts = df['payment_method'].value_counts()

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
payment_method_counts.plot(kind='bar', color='skyblue')
plt.title('Frecuencia de Métodos de Pago')
plt.xlabel('Método de Pago')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
````

# Ejemplo 2: Gráficas de línea
Para visualizar la compra más económica a la más costosa, podemos crear una gráfica de línea que muestre los valores de `discounted_total` ordenados de menor a mayor. A continuación, se muestra el código para generar esta gráfica:


````python
# Ordenar los valores de discounted_total de menor a mayor
sorted_discounted_total = df['discounted_total'].sort_values()

# Crear la gráfica de línea
plt.figure(figsize=(10, 6))
plt.plot(sorted_discounted_total.values, color='blue')
plt.title('Compras de Menor a Mayor Costo')
plt.xlabel('Índice')
plt.ylabel('Total con Descuento')
plt.show()
````
